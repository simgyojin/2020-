```
get_admin_code = re.compile('\d+')  # 행정동 코드만 가져오기

admin_code = []
for i in range(len(year19)):
  for x in year19['행정구역']:
    admin_code.append(x)    

test_int=get_admin_code.findall(str(admin_code))

for i in range(len(year19)):
  year19.loc[i,'행정구역'] = int(test_int[i]) 
```
